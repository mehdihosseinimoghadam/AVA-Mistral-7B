```python
from models.search import Search
```

```python
S = Search(data_path='./data/job_titles.csv',
           hf_model_name='LABSE')
```

```python
data = S.prep_data()
data[:10]
```

## In the following cell I use  [Sentence-bert](https://www.sbert.net/) as a feature extractor, specifically I use LABSE Model which is more accurate for more info on model take look at this [Model Hub](https://www.sbert.net/docs/pretrained_models.html#sentence-embedding-models)

### `SentenceTransformer.encode` method will encode each data point and will extract its embedding which later can be used both for semantic search and RAG applications with LLMs

```python
embeddings = S.create_engine(data)
```

#Search Function
### This Function will take query sentence `input_question` and number of relevent results that we want to have `num_res` as input then it will encode the query with the same method. Then as a vectors in a vector database it will find the `#num_res` vectors that are similar to our query with different similarity metrics.
![Sonny and Mariel high fiving.](https://raw.githubusercontent.com/UKPLab/sentence-transformers/master/docs/img/SemanticSearch.png)

```python
df = S.search("I am looking for machine learning with £80K range in Oxford or London",
              5,
              data,
              embeddings)
df
```

```python
from models.mistral_rag import MistralRAG
```

```python
RAG = MistralRAG('NousResearch/Nous-Hermes-2-Mistral-7B-DPO')
```

```python
job_des = 'I am looking for machine learning with £80K range in Oxford or London'

res = S.search("I am looking for machine learning with £80K range in Oxford or London",
              10,
              data,
              embeddings)


prompt = f'''

You are an assistant which help users to find best job match according to their specification from context provided,
remeber you are a freindly assistant which must answer users query accurately.

User Query:{job_des},

according to job titles in context which is the best match for users specifications

context: {res['res'][0]} - {res['res'][1]} - {res['res'][2]} - {res['res'][3]} - {res['res'][4]} - {res['res'][5]} - {res['res'][6]} - {res['res'][7]} - {res['res'][8]} - {res['res'][9]}

'''


prompt = f"### Human:{prompt}\n### Assistant:"



RAG.query(prompt, 500)
```
